## Net Approx. Algorithm

```python
# the NetApprox algorithm of MR22.
# 

'''
For fixed lambda, compute maximum weight of any weight vector w
that has lambda in OLS(X,y,w)
'''
def solve_fixed_lambda(X,XR,lam):
    n = X.shape[0]
    d = X.shape[1]
    c = -np.ones((n))
    A_eq = np.zeros((d,n))
    b_eq = np.zeros((d))
    for i in range(d):
        A_eq[i] = (X.T)[i] * (XR@lam)
    m = gp.Model("net-model")
    m.Params.OutputFlag = 0
    w = m.addMVar(shape=n,vtype=GRB.CONTINUOUS,name="w",lb=0,ub=1)
    m.setObjective(c @ w, GRB.MINIMIZE)
    m.addConstr(A_eq @ w == b_eq)
    m.optimize()
    return w.X

'''
NET UPPER BOUND

Input:
- X: n x d (covariate matrix)
- y: n (response vector)
- trials: integer

Output:
- Net upper bound on Stability(X,y)
'''
def net_algorithm(X,y,trials):
    n = X.shape[0]
    d = X.shape[1]
    XR = np.zeros((d,n))
    for i in range(d-1):
        XR[i] = (X.T)[i+1]
    XR[d-1] = y
    XR = XR.T
    assert(np.linalg.matrix_rank(XR)==d)
    U,s,Vh = scipy.linalg.svd(XR, full_matrices=False)
    A = Vh.T @ np.diag(1.0/s) @ np.eye(d)
    wstar = np.zeros((n))
    for i in range(trials):
        v = np.random.multivariate_normal(np.zeros((d)),np.eye(d))
        v = v / np.linalg.norm(v)
        lam = A@v
        w = solve_fixed_lambda(X,XR,lam)
        if sum(w) > sum(wstar):
            wstar = w
    return n-sum(wstar), wstar
```

```python
start_time = time.time()
n_sumwstar, wstar = net_algorithm(x,y,1000)
end_time = time.time()
print(f"N = {total_observations}, d = {d}, Stability = {n_sumwstar}, Time = {end_time - start_time}")
# Feb 12th: >10 min (for N=5000, d=200)
# Feb 12th: 3.42 min (for N=5000, d=50)
```

## Greedy Code: Lighter Rendition.

```python
def greedy_light(x, y, total_observations, method='IF'):
    lr = LinearRegression(x=x.T, y=y)
    lr.fit()
    beta = lr.beta[0] # pick an arbitrary dimension of beta as the QoI.
    floor_alphaN = np.floor(total_observations * .01).astype(int)
    for __ in range(floor_alphaN):
        # compute the one-step newton update to the full model for dropping the ith observation.
        if method == 'IF':
            if_scores = -lr.influence_scores()[0]
            sorted_if_scores = np.sort(if_scores)
            largest_if_score = sorted_if_scores[0]
            beta += largest_if_score
            drop_index = np.argmax(np.abs(if_scores))
        else:
            newton_scores = -lr.one_step_newton()[0]
            sorted_newton_scores = np.sort(newton_scores)
            largest_newton_score = sorted_newton_scores[0]
            beta += largest_newton_score
            drop_index = np.argmax(np.abs(newton_scores))
        
        x = np.delete(x, drop_index, axis=0)
        y = np.delete(y, drop_index)
        # update the linear regression model.
        lr = LinearRegression(x=x.T, y=y)
        lr.fit()
    return beta
```

```python
# Record runtimes.
amip_times = []
add1Exact_times = []
mis_size = []
# Set dimensions.
dimension_list = [2, 5, 10, 25, 50] # , 5, 10, 25, 50
# Maximum number of observations to remove.
alphaN = int(total_observations*0.01)

for elt in dimension_list:
    # Set dimensions.
    total_observations = 75000
    d = elt

    # Generate X, a d-dimensional isotropic Gaussian.
    mean = np.zeros(d)
    cov = np.identity(d)
    x = np.random.multivariate_normal(mean, cov, total_observations)

    # Generate Y, a linear combination of x with some noise.
    sigma = 1
    epsilon = np.random.normal(0, sigma, total_observations)
    y = []
    # take the inner product of x and beta and add epsilon.
    for i in range(total_observations):
        y.append(np.dot(x[i], np.ones(d)) + epsilon[i])
    y = np.array(y)
    print(f'X shape is {x.shape}, y shape is {y.shape}.' )
    
    lr = LinearRegression(x=x.T, y=y)
    lr.fit()
    print(
        f"Estimated beta1 with full dataset: {lr.beta[1]}\n"
    )
    
    # AMIP
    # start recording time.
    start_time = time.time()

    # compute scores.
    scores = -lr.influence_scores()[1]
    # sort.
    inds = np.argsort(scores)
    sorted_scores = scores[inds]
    sorted_if_scores = sorted_scores
    # sum top alpha N,
    sum(sorted_if_scores[:alphaN])

    # end recording time.
    end_time = time.time()

    amip_times.append(end_time - start_time)

    # Add1Exact
    # start recording time.
    start_time = time.time()
    # compute scores.
    newton_scores = -lr.one_step_newton()[1]
    # sort.
    inds = np.argsort(newton_scores)
    sorted_scores = newton_scores[inds]
    sorted_newton_scores = sorted_scores
    # sum top alpha N,
    sum(sorted_newton_scores[:alphaN])

    # end recording time.
    end_time = time.time()
    add1Exact_times.append(end_time - start_time)
    print("done dimension: ", d)
```
